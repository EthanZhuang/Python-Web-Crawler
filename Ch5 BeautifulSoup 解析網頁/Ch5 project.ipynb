## 1. 爬蟲實戰，圖片下載


````python
# ch5_14.py
import bs4, requests, os

url = 'http://www.grandtech.info/'                  # 上奇資訊網頁
html = requests.get(url)

try:
    html.raise_for_status()                             # 驗證網頁是否下載成功                      
    print("Download Success")
except HTTPError as e: 
    print(e)
    
destDir = 'out5_14'                                 # 設定未來儲存圖片的資料夾
if os.path.exists(destDir) == False:
    os.mkdir(destDir)                               # 建立資料夾供未來儲存圖片

objSoup = bs4.BeautifulSoup(html.text, 'lxml')      # 建立BeautifulSoup物件

imgTag = objSoup.select('img')                      # 搜尋所有圖片檔案
print("搜尋到的圖片數量 = ", len(imgTag))            # 列出搜尋到的圖片數量
if len(imgTag) > 0:                                 # 如果有找到圖片則執行下載與儲存
    for i in range(len(imgTag)):                    # 迴圈下載圖片與儲存
        imgUrl = imgTag[i].get('src')               # 取得圖片的路徑
        
        # 會有圖片開頭是 'http'
        if imgUrl.startswith('http'):
            continue
        
        print("%s 爬蟲到的 img ... \n" % imgUrl)
        finUrl = url + imgUrl                       # 取得圖片在Internet上的路徑
        print("%s 圖片下載中 ... \n" % finUrl)
        picture = requests.get(finUrl)              # 下載圖片
        
        picture.raise_for_status()                  # 驗證圖片是否下載成功
        print("%s 圖片下載成功" % finUrl)

        # 先開啟檔案, 再儲存圖片
        # 因為有 os.path.basement(imgUrl)，回傳是 out5_14\sample.jpg 
        pictFile = open(os.path.join(destDir, os.path.basename(imgUrl)), 'wb')
        for diskStorage in picture.iter_content(10240):
            pictFile.write(diskStorage)
        pictFile.close()                            # 關閉檔案
    
````

## 2. 下載圖片於防爬蟲軟件的網站

````python
# ch5_15.py
import bs4, requests, os
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg

# 使用 header 來偽裝瀏覽器
headers = { 'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64)\
            AppleWebKit/537.36 (KHTML, like Gecko) Chrome/45.0.2454.101\
            Safari/537.36', }

# 下載網頁內容 (此伺服器會擋住網頁)
url = 'http://aaa.24ht.com.tw/'                         
html = requests.get(url, headers = headers)           
print("網頁下載中 ...")

# 確認有無下載成功
try:
    html.raise_for_status()                             # 驗證網頁是否下載成功                      
    print("Download Success")
    print('-' * 70)
except HTTPError as e: 
    print(e)

# 設定儲存資料夾
destDir = 'out5_15'        
# 確認有無此資料夾，若沒有則建立目錄供未來儲存圖片
if os.path.exists(destDir) == False:
    os.mkdir(destDir)                               

# 建立BeautifulSoup物件
objSoup = bs4.BeautifulSoup(html.text, 'lxml')      

imgTag = objSoup.select('img')                      # 搜尋所有圖片檔案
print("搜尋到的圖片數量 = ", len(imgTag))            # 列出搜尋到的圖片數量
if len(imgTag) > 0:                                 # 如果有找到圖片則執行下載與儲存
    for i in range(len(imgTag)):                    # 迴圈下載圖片與儲存
        imgUrl = imgTag[i].get('src')               # 取得圖片的路徑
        print("%s 圖片下載中 ... " % imgUrl)
        
        # 取得圖片在Internet上的路徑 (finUrl)
        finUrl = url + imgUrl                       
        print("%s 圖片下載中 ... " % finUrl)
        # 下載圖片
        picture = requests.get(finUrl, headers = headers) 
        
        # 驗證圖片是否下載成功
        try:
            picture.raise_for_status()                     
            print("%s 圖片下載成功" % finUrl)            
        except HTTPError as e: 
            print(e)

        # 先開啟資料夾(destDir), 再儲存圖片
        # 因為有 os.path.basement(imgUrl)，回傳是 out5_14\sample.jpg 
        pictFile = open(os.path.join(destDir, os.path.basename(imgUrl)), 'wb')
        # 這裡的 iter_content(10240) 是利用跌代下載的方式，一秒下載 10240 bytes
        for diskStorage in picture.iter_content(10240):
        # 圖片寫入 pitcFile 內
            pictFile.write(diskStorage)
        pictFile.close()                            # 關閉檔案
        
````

````python
import bs4, requests

# 樂透網址
url = 'https://www.taiwanlottery.com.tw/'
# 要求取得網站資訊
html = requests.get(url)
print('downloading...')

# 確認下載網頁資料是否成功
if html.status_code == 200:
    print('Successful download')

# 利用 bs4 lxml 方法解析，並存入 bs4 物件
objSoup = bs4.BeautifulSoup(html.text, 'lxml')
# 在 bs4 物件中找所有 CSS class 屬性為 content_box02 的元素
dataTag = objSoup.select('.contents_box02')
# 確認總共有幾個
print('串列長度: ', len(dataTag))
# 先篩選出大樂透的中獎號碼
for i in range(len(dataTag)):
    print(dataTag[i])

# 樂透號碼為 Tag 為 div 且 class = 'ball_tx ball_green' 
# 找尋開出順序與大小順序的球
balls = dataTag[0].find_all('div', {'class': 'ball_tx ball_green'})
print('\n開出順序: ', end = '')
# 因為一次有12個號碼(ball_tx ball_green) + 1個特別號(ball_red) 
for i in range(int(len(balls)/2)):
    print(balls[i].text, end = '  ')

print('\n開出大小: ', end='')
for i in range(int(len(balls)/2), len(balls)):
    print(balls[i].getText(), end = '  ')

# 找出第二區的紅球    
redball = dataTag[0].find_all('div', class_ = 'ball_red')
print('\n特別號: ', redball[0].text)
````
