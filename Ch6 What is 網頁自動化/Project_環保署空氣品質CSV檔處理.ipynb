## 1. 讀取環保署空氣品質 CSV file，而列出城市名稱、站台名稱、站台ID、PM2.5值

```python
import csv
import pandas as pd

infn_old = 'AQI_20190814010150.csv'
infn_new = 'AQI_20200911155434.csv'
outfn = 'out6_12.csv'

f = open(infn_old)
data = pd.read_csv(f)
print(data)

# 讀取檔案，開啟檔案
with open(infn_old) as old_csvFile:
    old_csvReader = csv.reader(old_csvFile)
    old_listReport = list(old_csvReader)
# print(old_listReport)

# 列出程式名稱
old_newlistReport = list() 
tmplist = []

for row in old_listReport:
#     print()
    tmplist = [row[1], row[23], row[11], row[0]]
    old_newlistReport.append(tmplist)
    
with open(outfn, 'w', newline = '') as csvout:
    csvWriter = csv.writer(csvout)
    for row in old_newlistReport:
        csvWriter.writerow(row)
        if row[0] != 'County':
            print('城市名稱 =%4s\t站台ID =%3s\tPM2.5值 =%3s\t站台名稱 =%s ' \
                  %(row[0], row[1], row[2], row[3]))
```
